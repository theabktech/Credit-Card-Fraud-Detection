```python
import h2o
from h2o.estimators import H2OXGBoostEstimator
h2o.init()

# Import the titanic dataset into H2O:
dataset = h2o.import_file("/content/drive/MyDrive/Credit Card Fraud Dataset/creditcard.csv")

# Set the predictors and response; set the response as a factor:
dataset["Class"] = dataset["Class"].asfactor()
predictors = dataset.columns
response = "Class"

# Split the dataset into a train and valid set:
train, valid = dataset.split_frame(ratios=[.8], seed=1234)

# Build and train the model:
ccfd_xgb = H2OXGBoostEstimator(booster='dart',
                                  normalize_type="tree",
                                  seed=1234)
ccfd_xgb.train(x=predictors,
                  y=response,
                  training_frame=train,
                  validation_frame=valid)

# Eval performance:
perf = ccfd_xgb.model_performance()

# Generate predictions on a test set (if necessary):
pred = ccfd_xgb.predict(valid)

# Extract feature interactions:
feature_interactions = ccfd_xgb.feature_interaction()
```
